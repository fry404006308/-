### 批量显示图片

```python
# 从begin的位置开始，显示step张图
def show_img(begin=0,step=25):
    figure = plt.figure(figsize=(10,10))
    for i in range(step):
        plt.subplot(5, 5, i + 1, title='name')
        plt.xticks([])
        plt.yticks([])
        plt.imshow(train_x[i+begin])
    return figure,
```

```python
show_img()
```

```python
show_img(10,8)
```

```python
show_img(100,12)
```
